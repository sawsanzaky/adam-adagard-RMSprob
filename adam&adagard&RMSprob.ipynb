```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,20,50)  # to divide space 0 to 20 as 50 chruck
x
```

```python
y = np.array((-2*x)+1)
y
```

```python
real_data=pd.DataFrame({'x':x,'y:=yi = a*xi + b (and a = -2, b = 1)':y})
real_data
```

# Adagard 

```python
vt_0 , vt_1 = 0,0
alpha=0.001
epsilon = 0.0001
max_iteration =50
theta0 , theta1 = 1,1
tolarance = 0.001
list_theta0=[]
list_theta1=[]
list_loss=[]
list_vt0=[]
list_vt1=[]
```

```python
for i in range (max_iteration):
    h_x=theta0+theta1*x
    vt_0 = vt_0 + pow((sum(h_x-y)/len(y)),2)
    list_vt0.append(vt_0)
    theta0=theta0 - ((alpha*sum(h_x-y)/len(y))/(pow(vt_0,0.5)+epsilon))
    list_theta0.append(theta0)
    vt_1 = vt_1 + pow((sum((h_x-y)*x)/len(y)),2)
    list_vt1.append(vt_1)
    theta1 = theta1 -((alpha*sum((h_x-y)*x)/len(y))/(pow(vt_1,0.5)+epsilon))
    list_theta1.append(theta1)
    loss = sum((h_x-y)**2)/(2*len(y))
    list_loss.append(loss)
    plt.scatter(loss,theta0)
    if loss <= tolarance :
        break
plt.xlabel('loss')
plt.ylabel('theta0')
plt.title(" loss vs theta0")
```

```python
data_show = pd.DataFrame({'theta0':list_theta0,'theta1':list_theta1,'loss':list_loss,'vt_0':list_vt0,'vt_1':list_vt1})
data_show
```

```python
#data_show.plot(x='theta0','theta1','loss','vt_0','vt_1')
data_show.plot(x='loss',y='vt_0')
plt.show()
data_show.plot(x='loss',y='vt_1')
plt.show()
data_show.plot(x='loss',y='theta0')
plt.show()
data_show.plot(x='loss',y='theta1')
plt.show()
```
